```python
import torch
import torch.optim as optim
```

```python
t_c = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0,
                    8.0, 3.0, -4.0, 6.0, 13.0, 21.0])
t_u = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9,
                    33.9, 21.8, 48.4, 60.4, 68.4])
t_un = 0.1 * t_u

def model(t_u, w, b):
    return w * t_u + b

def loss_fn(t_p, t_c):
    return ((t_p - t_c)**2).mean()

def training_loop(n_epochs,
                  optimizer,
                  params, t_u, t_c):
    
    for n_epoch in range(n_epochs):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if n_epoch % 500 == 0:
            print('Epoch %d, Loss %f' % (n_epoch, float(loss)))
    return params
    
# Initial parameters
params = torch.tensor([1.0, 0.0], requires_grad=True)
learning_rate = 1e-5
optimizer = optim.SGD([params], lr=learning_rate)

# Trying the training loop with SGD optimizer
training_loop(
    n_epochs=5000,
    optimizer=optimizer,
    params=params,
    t_u=t_un,
    t_c=t_c
)
```

```python
# Trying the 
params = torch.tensor([1.0, 0.0], requires_grad=True)
learning_rate = 1e-1
training_loop(
    n_epochs=2000,
    optimizer=optim.Adam([params], lr=learning_rate),
    params=params,
    t_u = t_u,
    t_c = t_c
)
```
